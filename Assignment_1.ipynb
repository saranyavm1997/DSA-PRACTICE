**5. Perform meaningful visualizations using the dataset.**

```python
sns.boxplot(data)
```

```python
plt.bar(data['Classification'],data['SL'], width = 0.5)
plt.xlabel = "Classification"
plt.ylabel = "SL"
plt.show()
```

```python
sns.countplot(x= 'Classification', hue = 'Classification', data = data,palette='Accent', legend = False)
```

```python
data["Classification"].value_counts()
```

```python
classification_data = data.groupby(by = 'Classification').sum(numeric_only = True)
classification_data
```

```python
classification_data[["SL", "SW", "PL", "PW"]].plot.bar()
plt.xticks(rotation = 0)
plt.title("Iris Classification")
```
